#1. Análisis de facturación

# 2. Ventas por categoría

```python
lista_tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df in enumerate(lista_tiendas, 1):
    # 1. Estandarizamos nombres (minúsculas y sin espacios extra)
    df.columns = df.columns.str.strip().str.lower()

    columna_objetivo = 'categoría del producto'

    if columna_objetivo in df.columns:
        print(f"--- 🏪 Tienda {i}: Top Categorías ---")
        # Agrupamos y contamos
        conteo = df.groupby(columna_objetivo).size().sort_values(ascending=False)
        print(conteo.head(5))
    else:
        print(f"❌ Sigo sin encontrar la columna en la Tienda {i}")

    print("\n" + "-"*30 + "\n")
```

# 3. Calificación promedio de la tienda


```python
lista_tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df in enumerate(lista_tiendas, 1):

    df.columns = df.columns.str.strip().str.lower()

    columna_calif = 'calificación'

    if columna_calif in df.columns:
        # Calculamos el promedio
        promedio = df[columna_calif].mean()

        # Contamos cuántas reseñas hay (para dar contexto al promedio)
        total_resenas = df[columna_calif].count()

        print(f"--- 🏪 Tienda {i} ---")
        print(f"Calificación Promedio: {promedio:.2f} ⭐")
        print(f"Basado en {total_resenas} evaluaciones.")
    else:
        print(f"❌ No encontré la columna '{columna_calif}' en la Tienda {i}")
    print("\n")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

lista_tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df in enumerate(lista_tiendas, 1):
    # Aseguramos limpieza de nombres
    df.columns = df.columns.str.strip().str.lower()

    # Contamos ventas por producto
    conteo_productos = df['producto'].value_counts()

    # Seleccionamos los 5 mejores y 5 peores
    top_5 = conteo_productos.head(5)
    bottom_5 = conteo_productos.tail(5)

    # --- Creación de la Gráfica ---
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    fig.suptitle(f'🏪 Análisis de Ventas - Tienda {i}', fontsize=16)

    # Gráfica de los Más Vendidos
    top_5.plot(kind='bar', ax=ax[0], color='skyblue')
    ax[0].set_title('Top 5 Más Vendidos')
    ax[0].set_ylabel('Cantidad de Ventas')

    # Gráfica de los Menos Vendidos
    bottom_5.plot(kind='bar', ax=ax[1], color='salmon')
    ax[1].set_title('5 Menos Vendidos')
    ax[1].set_ylabel('Cantidad de Ventas')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Preparación de datos consolidados
tiendas_nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
lista_df = [tienda, tienda2, tienda3, tienda4]

# Listas para almacenar métricas
ingresos_puros = []
costos_envio_totales = []
malas_calificaciones = [] # Conteo de 1 y 2 estrellas
ratios_eficiencia = []

for df in lista_df:
    df.columns = df.columns.str.strip().str.lower()

    ingreso = df['precio'].sum()
    costo = df['costo de envío'].sum()

    ingresos_puros.append(ingreso)
    costos_envio_totales.append(costo)

    # Contamos clientes insatisfechos (calificación <= 2)
    malas = len(df[df['calificación'] <= 2])
    malas_calificaciones.append(malas)

    # Ratio: Por cada dólar de envío, cuánto ingresa
    ratios_eficiencia.append(ingreso / costo)

# --- CREACIÓN DEL DASHBOARD ---
fig, ax = plt.subplots(1, 3, figsize=(22, 6))
fig.suptitle('Análisis de Ineficiencia: ¿Qué tienda debe vender el Sr. Juan?', fontsize=18, fontweight='bold')

# GRAFICO 1: Ingresos vs Costos (Barras Apiladas)
ax[0].bar(tiendas_nombres, ingresos_puros, label='Ingreso Neto', color='seagreen')
ax[0].bar(tiendas_nombres, costos_envio_totales, label='Costo Envío', color='tomato', bottom=0) # Visualmente comparativo
ax[0].set_title('Impacto del Envío en el Ingreso')
ax[0].legend()

# GRAFICO 2: Clientes Insatisfechos (Puntos Críticos)
barras_malas = ax[1].bar(tiendas_nombres, malas_calificaciones, color='firebrick')
ax[1].bar_label(barras_malas, padding=3, fontweight='bold')
ax[1].set_title('Cantidad de Clientes Insatisfechos (1-2 ⭐)')
ax[1].set_ylabel('Número de Reseñas Negativas')

# GRAFICO 3: Ratio de Eficiencia (El factor decisivo)
# Cuanto más bajo el ratio, más ineficiente es la tienda
colores_ratio = ['orange' if r == min(ratios_eficiencia) else 'lightgray' for r in ratios_eficiencia]
barras_ratio = ax[2].bar(tiendas_nombres, ratios_eficiencia, color=colores_ratio)
ax[2].bar_label(barras_ratio, fmt='%.2f', label_type='center', color='black', fontweight='bold', rotation=90)
ax[2].set_title('Ratio de Eficiencia (Ingreso / Envío)')
ax[2].set_ylabel('Retorno por cada $1 de envío')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# 7. Informe



1. Introducción 🎯
El objetivo de este análisis es identificar cuál de las cuatro tiendas de la cadena presenta el rendimiento más bajo o la menor eficiencia operativa. A través del análisis de KPIs (Indicadores Clave de Desempeño), determinaremos qué unidad de negocio debe ser vendida para financiar el nuevo proyecto del Sr. Juan, minimizando el impacto negativo en el flujo de caja total de la empresa.

2. Análisis de Factores Críticos 📊
A continuación, sintetizamos los hallazgos basados en los datos procesados:

Rendimiento Financiero (Ingresos): Aunque todas las tiendas operan en rangos similares, la Tienda [4] muestra un rezago frente a la Tienda 1. Una menor facturación sugiere una saturación de mercado o una demanda estancada.

Eficiencia Logística (Costos de Envío): Hemos detectado que el costo de envío promedio es un factor determinante. Si una tienda tiene costos de envío altos y ventas bajas, su margen de competitividad es menor.

Satisfacción del Cliente (Rating): Un promedio de calificación bajo (cercano a 3 o menos) es una "red flag". Indica que mantener esa tienda requiere una inversión alta en servicio al cliente o gestión de devoluciones.



1. La Paradoja de la Tienda 1: "Mucho Brillo, Poco Filtro" 🚨
Es muy curioso: la Tienda 1 es la que más dinero ingresa ($1,150M), pero  es también la que tiene más clientes insatisfechos (405 reseñas negativas) y el menor Ratio de Eficiencia (18.75).

**Análisis:** Ser el más grande no significa ser el mejor. La Tienda 1 parece estar "muriendo de éxito". Tiene tanto volumen que su logística de envíos es menos eficiente y está acumulando más quejas que las demás.

2. Eficiencia Casi Idéntica
 las cuatro tiendas están prácticamente empatadas (18.75 vs 18.77).

**Análisis:** Esto significa que el modelo de negocio es muy estable. No hay una tienda que sea un "desastre" total, lo cual es bueno, pero hace la decisión de vender más difícil. El Sr. Juan no venderá una tienda "rota", sino una que simplemente le da un poco menos de retorno por cada dólar invertido.

3. El Motor del Negocio: Muebles y Electrónicos
El gráfico de tarta es contundente: el 42.1% de los ingresos vienen solo de dos categorías.

**Análisis:** Cualquier decisión sobre qué tienda vender debe considerar si esa tienda es fuerte en estas categorías. Si la Tienda 4 (que es la que menos ingresa en total) también es la más débil en Muebles, sería la candidata ideal para salir.

4. Relación Precio-Satisfacción: El "Techo" de las 5 Estrellas ⭐
 Hay una gran densidad de productos de bajo costo con 5 estrellas, pero a medida que el precio sube (hacia los 3 millones), las calificaciones se mantienen altas.

**Análisis:** No hay una correlación negativa; es decir, la gente no se queja más solo porque el producto sea más caro. Esto sugiere que la calidad de los productos de alto ticket es buena.
